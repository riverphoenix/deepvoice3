```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
from matplotlib.pyplot import step, show
import librosa.display
import pyworld as pw
import IPython

%matplotlib inline  

sound_file = "datasets/LJMag/wavs/LJ001-0012.wav"
sr=22050
fm = 5
n_fft = 2048 # fft points (samples)
frame_shift = 0.0125 # seconds 0.0125
frame_length = 0.05 # seconds 0.05
hop_length = int(sr*frame_shift) # samples  This is dependent on the frame_shift.
win_length = int(sr*frame_length) # samples This is dependent on the frame_length.
n_mels = 80 # Number of Mel banks to generate
preemphasis = 0.97 

y, sr = librosa.load(sound_file, sr)
y, _ = librosa.effects.trim(y)
y_pre = np.append(y[0], y[1:] - preemphasis * y[:-1])
linear = librosa.stft(y=y_pre,
                      n_fft=n_fft,
                      hop_length=hop_length,
                      win_length=win_length)
mag = np.abs(linear)  # (1+n_fft//2, T)
mel_basis = librosa.filters.mel(sr,n_fft, n_mels)  # (n_mels, 1+n_fft//2)
mel = np.dot(mel_basis, mag)  # (n_mels, t)
mel = librosa.amplitude_to_db(mel)
mel = mel.T.astype(np.float32)  # (T, n_mels)

#pitch, harmonic, aperiodic = pw.wav2world(np.float64(y), sr,frame_period=fm)
pitch, harmonic, aperiodic = pw.wav2world(np.float64(y), sr)
pitch = pitch.astype(np.float32)
harmonic = harmonic.astype(np.float32)
aperiodic = aperiodic.astype(np.float32)

plt.figure(figsize=(10, 10))

ind = np.arange(len(pitch))
width = 0.35

ax = plt.subplot(5, 1, 1)
ax.bar(ind, pitch, width, color='r')
plt.title('Predicted Pitch')
plt.tight_layout()

plt.subplot(5, 1, 2)
librosa.display.specshow(harmonic.T,y_axis='log')
plt.title('Predicted harmonic')
plt.colorbar()
plt.tight_layout()

plt.subplot(5, 1, 3)
librosa.display.specshow(aperiodic.T)
plt.title('Predicted Aperiodic')
plt.colorbar()
plt.tight_layout()

print(len(pitch))
print(harmonic.shape)
print(aperiodic.shape)
print(mel.shape)

#y_out = pw.synthesize(np.float64(pitch), np.float64(harmonic), np.float64(aperiodic), sr,frame_period=fm)
y_out = pw.synthesize(np.float64(pitch), np.float64(harmonic), np.float64(aperiodic), sr)

plt.subplot(5,1,4)
librosa.display.waveplot(y, sr=sr)
plt.title("Original")

plt.subplot(5,1,5)
librosa.display.waveplot(y_out, sr=sr)
plt.title("Synth")

plt.show()

IPython.display.display(IPython.display.Audio(y,rate=sr))
IPython.display.display(IPython.display.Audio(y_out,rate=sr))
```
